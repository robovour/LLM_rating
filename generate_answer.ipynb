```python
import pandas as pd
import numpy as np
import logging, json, requests
logger = logging.getLogger(__name__)
```

```python
def _post(data:dict) ->str:
    try:
        res = requests.post("http://127.0.0.1:8000", json.dumps(data), headers={"Content-Type":"application/json"})
        if res.status_code == 200:
            output = json.loads(res.text)
            #self.history += [(data['prompt'],output['response'])]
            return output['response']
        else:
            logger.error("Error when connect to server.")
            return "No output"
    except Exception as e:
        logger.exception(e)
```

```python
df = pd.read_excel("医疗/cMedQA2.xlsx")
df.insert(df.shape[1],'response',"None")
for index,row in df.iterrows():
    data = {
        "prompt":'现在你是一位金融顾问，请对我之后的提问进行简短回答，请注意控制字数不要超过50字。\n'+row['question'],
        "history":[]
    }
    df.loc[index,'response'] = _post(data)
df
```

```python
df.to_excel("医疗/cMedQA2.xlsx")
```
