```python
import fastapi_poe as fp
import pandas as pd
```

```python
df = pd.read_excel('第二轮处理数据/医疗/pqal_100.xlsx')
```

```python
df["rating_form"] = "[question]\n" + df["question"] + "\n[response]\n" + df["response"] + "\n[referense answer]\n" + df["answer"]
```

```python
df.insert(df.shape[1],"GPT4_rating",0)
df
```

```python
for index,row in df.iterrows():
    system_message = fp.ProtocolMessage(role="system", content="""You are a fair judge for the quality of responses generated by large language models. According to the given question, response and reference answer, rate the response to the question on a scale of 1-10, and provide the reasons for your rating. You must strictly follow the format:"[[rating]];{{reasons}}" """)
    question_message = fp.ProtocolMessage(role='user',content=df.loc[index,"rating_form"])
    respond = ''
    async for partial in fp.get_bot_response(messages=[system_message,question_message], bot_name="GPT-4", api_key='4ln7XNGyuQ6f_YrryRMEFj5DZHfIdnhktnj3RcMj_Cw'): 
        respond += partial.text
    df.loc[index,"GPT4_rating"] = respond
```

```python
import json
```

```python
df = pd.read_excel('第二轮处理数据/医疗/pqal_100.xlsx')
```

```python
qa = []
for index,row in df.iterrows():
    conversations = []
    talk = {}
    talk["role"] = "user" 
    talk["content"] = """You are a fair judge for the quality of responses generated by large language models. According to the given question, response and reference answer, rate the response to the question on a scale of 1-10, and provide the reasons for your rating. You must strictly follow the format:"[[rating]];{{reasons}}" """\
        + row["rating_form"]
    conversations.append(talk.copy())
    talk["role"] = "assistant"
    talk["content"] = row["GPT4_rating"]
    conversations.append(talk.copy())
    qa.append(conversations.copy())
with open("第二轮处理数据/医疗/pqal_100.json","w",encoding="utf-8") as f:
    json.dump(qa,f,ensure_ascii=False)
```
